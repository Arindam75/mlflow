```python
import mlflow
import mlflow.keras
import os
```

Loading data pertaining to an experiment

```python
mlflow.set_tracking_uri("sqlite:///mlflow.db")
mlflow.search_runs(experiment_names = ["keras_log"])
```

Using any of the run_ids we can load a model

```python
run_relative_path_to_model="model"
run_id = "194dff3ecf6e4a8089316769478a7c55"
model_uri = f"runs:/{run_id}/{run_relative_path_to_model}"
```

```python
model = mlflow.keras.load_model(model_uri)

model.summary()
```
